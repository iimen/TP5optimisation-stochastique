```python
import numpy as np

#Fonction a optimiser 
def LeonardJoues(x):
  f = 0
  n=x.shape[0]
  for i in range (n-1):
    for j in range (i+1, n):
        f += (1)/(x[i] - x[j])**12 - (1)/(x[i] - x[j])**6
    return f
```

```python
#Dimension 8
from geneticalgorithm import geneticalgorithm as ga

ndim = 8
varbound=np.array([[-2,2]]*ndim*3)

parametres = {'max_num_iteration': 350,\
                  'population_size':500,\
                  'mutation_probability':0.1,\
                  'elit_ratio': 0.022,\
                  'crossover_probability': 0.5,\
                  'parents_portion': 0.3,\
                  'crossover_type':'uniform',\
                  'max_iteration_without_improv':None}

model=ga(function=LeonardJoues,dimension=ndim*3,variable_type='real',variable_boundaries=varbound,algorithm_parameters=parametres)
model.run()
```
